**How the normalized score are calculated**

$$\mathbf{\hat{X}}_i = \frac{\textnormal{X}_i - \displaystyle\min_{\textnormal{X}}}{\displaystyle\max_{\textnormal{X}}-\displaystyle\min_{\textnormal{X}}}$$

---

**How the overall score are calculated**

given $s$ as the semantic similarity score (value given by text embedding), $k$ as the skills similarity score, $ko$ as the keyword overlap score, $e$ as the entity match score, the overall score are calculated with the equation:

$$ \textnormal{raw} = s\times w_s + k \times w_k + ko \times w_{ko} + e \times w_e $$

where $w$ denotes each weight set by the user. for the context of this thesis, it used $w_s = 0.05$, $w_k = 0.7$, $w_{ko} = 0.2$, $w_e = 0.05$ and $w_s+w_k+w_{ko}+w_e = 1$

later, the $raw$ value can be taken as the input for score normalization equation mentioned prior

## Descriptive Statistic for each score

## Histogram for each metrics

```python
for col in cols:
  plt.figure(figsize=(10, 5))
  sns.histplot(df[col], kde=True)
  plt.title(f'Histogram for {col}')
  plt.show()
```
